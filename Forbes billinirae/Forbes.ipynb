```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(25, 20), dpi=200, facecolor='w', edgecolor='k')

df["age"].value_counts(ascending=True).plot.barh(color=["#cdb4db", "#ffc8dd", "#ffafcc"])

# Description
plt.title("Age X Frequency")
plt.ylabel("Age")
plt.xlabel("Frequency")

# Save the Figure
plt.savefig('age_distribution.png')
plt.show()
```

```python
plt.figure(figsize=(26,20))
sns.countplot(data=df,x='age',palette='rocket')
plt.title('Age X Frequency',fontsize=25)
plt.show()
```

```python
plt.figure(figsize=(80,20),dpi=200)
sns.countplot(data=df,x='networth',palette='viridis')
plt.title('Net worth X Frequency')
plt.savefig('Networth_frequency.png')
plt.show()
```

```python
plt.figure(figsize=(60,30),dpi=200)
sns.countplot(data=df,x='industry',palette='Greens')
plt.title('Industry X frequency')
plt.savefig('industry_frequency.png')
plt.show()
```

```python
plt.figure(figsize=(100,30),dpi=200)
sns.catplot(data=df,x='age',y='industry',kind='bar')
plt.title('Age vs Industry')
```
