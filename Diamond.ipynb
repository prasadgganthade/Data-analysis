#### Find the cut unique value

```python
df.cut.value_counts()
```

```python
sns.countplot(data=df,x='cut')
plt.title('Frequency X Cut')
plt.show()
```

```python
df.cut.value_counts().plot(kind='pie',autopct='%1.2f%%')
plt.title('Distribution of cut')
```

From above Ideal has highest quantity and falir has lowest quantity

#### Find Unique value in color

```python
df.color.value_counts()
```

```python
sns.countplot(data=df,x='color')
plt.title('Frequency X Color')
plt.show()
```

From above G color has highest count and j color has lowest count.

#### Find unique value in clarity

```python
df.clarity.value_counts()
```

```python
color = sns.color_palette('pastel')
df.clarity.value_counts().plot(kind='barh',color=color)
plt.title('Frequency X Clarity')
```

```python
sns.countplot(data=df,x='clarity',hue='cut')
plt.title('Clarity is hued by cut')
```

```python
sns.countplot(data=df,x='color',hue='cut')
plt.title('Color hued by cut')
```

```python
sns.boxplot(data=df,x='cut',y='carat')
```

```python
plt.figure(figsize=(30,12))
sns.catplot(data=df,x='table',y='carat',hue='cut')
```

```python
sns.pairplot(data=df)
```

```python
# Plotting price vs carat hued by cut
plt.figure(figsize=(20,14))
sns.scatterplot(data=df,x='price',y='carat',hue='cut')
plt.title('Relationship between price and carat hued by cut',fontsize=16)
```
