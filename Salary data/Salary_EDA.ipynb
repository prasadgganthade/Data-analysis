```python
plt.figure(figsize=(14,10))
sns.countplot(data=df,x='Location',hue='Employment_Status',palette='viridis')
plt.title('Count of Jobs in location wise and hued by employment status')
plt.show()
```

```python
sns.countplot(data=df,x='Employment_Status')
plt.title('Count of employment status')
plt.show()
```

```python
df.Employment_Status.value_counts().plot(kind='pie',autopct='%.f%%')
plt.title('% wise employment status')
plt.show()
```

```python
# What is the change of salary in different locations?
plt.figure(figsize=(16,10))
sns.barplot(data=df,x='Location',y='Salary',hue='Employment_Status')
sns.lineplot(data=df,x='Location',y='Salary',hue='Employment_Status',linewidth=2.5,marker='o')
plt.title('Salary of employment at different location hued by employment status',fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(16,10))
df.Job_Roles.value_counts().plot(kind='line',linewidth=2.5,marker='*',color='red',markersize=12)
df.Job_Roles.value_counts().plot(kind='bar',color=["#33032f", "#531253"])
plt.title('Frequency X Job roles',fontsize=16)
plt.show()
```

```python
df.Job_Roles.value_counts()
```

```python
sns.relplot(col='Job_Roles',data=df,y='Salary',x='Employment_Status',kind='line')
plt.show()
```

```python
plt.figure(figsize=(20,12))
figure_1 = sns.barplot(data=df,x='Location',y='Salary',palette='rocket')
figure_1 = sns.lineplot(data=df,x='Location',y='Salary',palette='viridis',marker='o',linewidth=2.5,markersize=12)
plt.title('Location vs Salary',fontsize=16)
plt.show()
```
