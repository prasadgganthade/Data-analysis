```python
new_col = pd.Series(df.columns).str.replace(" ","_")
```

```python
df.columns = new_col
```

```python
df.columns
```

```python
# Now creating new column like month order year and order day.
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
df['Ship_Date'] = pd.to_datetime(df['Ship_Date'])

df['Order_Month'] = df['Order_Date'].dt.month
df['Order_Year'] = df['Order_Date'].dt.year
df['Order_Day_of_Week'] = df['Order_Date'].dt.dayofweek
```

```python
# Now look the monthly sales
sales_by_month = df.groupby('Order_Month')['Sales'].sum().reset_index()
sales_by_month
```

```python
sns.barplot(data=sales_by_month,x='Order_Month',y='Sales')
plt.title('Monthly sales of mart')
plt.show()
```

```python
# Now by category sales
sales_by_category = df.groupby('Category')['Sales'].sum().reset_index()
sales_by_category
```

```python
sales_by_category.plot(kind='pie',y='Sales',autopct='%.02f%%')
plt.title('Sales by category')
```

Technology has highest sales compare to other category.

```python
# Now sales by sub-category
sale_by_sub_category = df.groupby('Sub-Category')['Sales'].sum().reset_index()
sale_by_sub_category
```

```python
plt.figure(figsize=(20,12))
sns.barplot(data=sale_by_sub_category,x='Sub-Category',y='Sales')
plt.title('Sales by sub category')
plt.show()
```

```python
# Now monthly profit
profit_by_month = df.groupby('Order_Month')['Profit'].sum().reset_index()
profit_by_month
```

```python
sns.lineplot(data=profit_by_month,x='Order_Month',y='Profit')
plt.title('Profit by month')
plt.show()
```

```python
plt.figure(figsize=(20,12))
sns.barplot(data=profit_by_subcategory,x='Sub-Category',y='Profit')
plt.title('Profit by sub category')
plt.show()
```

```python
sales_profit_by_segment = df.groupby('Segment').agg({'Sales':'sum','Profit':'sum'}).reset_index()
sales_profit_by_segment
```

```python
sns.lineplot(data=sales_profit_by_segment,x='Segment',y='Sales')
sns.lineplot(data=sales_profit_by_segment,x='Segment',y='Profit')
plt.show()
```
