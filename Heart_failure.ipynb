```python
# Load the data
heart = pd.read_csv('heart_failure_clinical_records_dataset.csv')
heart.head()
```

```python
# Printing last 5 rows
heart.tail()
```

### Some basic points exploring via pandas and numpy
1. **Anaemia:** 0 -> No || 1 -> Yes
2. **diabetes:** 0 -> No || 1 -> Yes
3. **high_blood_pressure:** 0 -> No || 1 -> Yes
4. **sex:** 0 -> Female || 1 -> Male
5. **smoking:** 0 -> No || 1 -> Yes
6. **DEATH_EVENT:** 0 -> No || 1 -> Yes

```python
df = pd.DataFrame()

df['age'] = heart['age']
df['anaemia'] = np.where(heart['anaemia'] < 1, 'No','Yes')
df['creatinine_phosphokinase'] = heart['creatinine_phosphokinase']
df['diabetes'] = np.where(heart['diabetes'] < 1, 'No','Yes')
df['ejection_fraction'] = heart['ejection_fraction']
df['high_blood_pressure'] = np.where(heart['high_blood_pressure'] < 1, 'No','Yes')
df['platelets'] = heart['platelets']
df['serum_creatinine'] = heart['serum_creatinine']
df['serum_sodium'] = heart['serum_sodium']
df['sex'] = np.where(heart['sex'] < 1, "Female",'Male')
df['smoking'] = np.where(heart['smoking'] < 1, 'No','Yes')
df['DEATH_EVENT'] = np.where(heart['DEATH_EVENT'] < 1, 'No','Yes')
```
